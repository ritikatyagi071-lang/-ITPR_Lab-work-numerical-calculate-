```
#Function to check the number is odd or even
def check_odd_even(num):
  if num%2==0:
    return "Even"
  else:
    return "Odd"

#Main function
def main():
  number=int(input("Enter a number:"))
  result=check_odd_even(number)
  print("The number is",result)

#Calling main function
main()
```
